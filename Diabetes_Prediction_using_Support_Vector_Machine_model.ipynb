Separating the data and labels

```python
X=Dataset.drop(columns='Outcome',axis=1) #axis is 1 means we are droping the column
Y=Dataset['Outcome']
```

```python
print(X)
print(Y)
```

Normalization of data so that we have the values in between 0,1 for better prediction

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
StandardScaler?
```

```python
Standardized_data=scaler.transform(X)
```

```python
Standardized_data
```

```python
X=Standardized_data
Y=Dataset['Outcome']
```

```python
print(X)
print(Y)
```

Splitting the data into training and testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Training The Model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#training the svm model
classifier.fit(X_train,Y_train)
```

MODEL EVALUATION

ACCURACY SCORE OF TRAINING DATA

```python
X_train_prediction=classifier.predict(X_train)
```

```python
X_train_prediction
```

```python
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy score of our training data is:",training_data_accuracy)
```

ACCURACY SCORE OF TESTING DATA

```python
X_test_prediction=classifier.predict(X_test)
```

```python
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score of testing data is:", testing_data_accuracy)
```

MAKING A PREDICTIVE SYSTEM

```python
input_data=(4,110,92,0,0,37.6,0.191,30)

input_data_as_numpy_array=np.asarray(input_data)

input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

std_data=scaler.transform(input_data_reshaped)

prediction=classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
  print("TheÂ personÂ isÂ notÂ diabetic")
else:
  print("TheÂ personÂ isÂ diabetic")
```

```python
input_data=(6,148,72,35,0,33.6,0.627,50)

input_data_as_numpy_array=np.asarray(input_data)

input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

std_data=scaler.transform(input_data_reshaped)

prediction=classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
  print("TheÂ personÂ isÂ notÂ diabetic")
else:
  print("TheÂ personÂ isÂ diabetic")
```

```python

# Function to predict diabetes
def predict_diabetes(input_data):
    input_data = [float(x) for x in input_data.split(',')]
    input_data_features = scaler.transform([input_data])

    prediction = classifier.predict(input_data_features)

    return "Non diabetic" if prediction[0] == 1 else "diabetic"

# CLI interface
if __name__ == "__main__":
    print("Welcome to the diabetes prediction system!")
    while True:
        user_input = input("Enter the patient record here (or type 'exit' to quit):\n")
        if user_input.lower() == 'exit':
            break
        result = predict_diabetes(user_input)
        print(f"Prediction: {result}\n")
```
